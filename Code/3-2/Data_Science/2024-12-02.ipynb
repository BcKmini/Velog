```r
# ggplot 객체 생성
p <- ggplot(data, aes(x = 번호, y = 공룡알화석산지))

# 점 추가
p <- p + geom_point(fill = 'red', color = 'blue', shape = 21, size = 4)

# 최종 플롯 출력
p + geom_line(color='blue', linetype=2)
```

```r
p <- ggplot(mtcars, aes(factor(cyl)))
p + geom_bar(fill='gold', color='red', linewidth=2, width=0.8)
```

```r
p <- ggplot(mtcars, aes(factor(cyl), fill=factor(gear)))
p + geom_bar()
```

```r
library(doBy)
tbl <- summaryBy(mpg ~ cyl, data=mtcars, FUN=mean)
tbl

p <- ggplot(tbl, aes(factor(cyl), mpg.mean))
p + geom_bar(stat='identity', fill='gold', color='red')
```

```r
library(doBy)
tbl <- summaryBy(mpg ~ cyl+gear, data=mtcars, FUN=mean)
tbl

p <- ggplot(tbl, aes(factor(cyl), mpg.mean, fill=factor(gear)))
p + geom_bar(stat='identity')
```

```r
#library(doBy)
#tbl <- summaryBy(mpg ~ cyl+gear, data=mtcars, FUN=mean)
#tbl

p <- ggplot(tbl, aes(factor(cyl), mpg.mean, fill=factor(gear)))
p + geom_col(position = 'dodge')
```

```r
library(dplyr)

data <- read.csv("../mydata/도로교통공단_시군구별 월별 교통사고 통계_20231231.csv",
                 fileEncoding="CP949")

chungju <- filter(data, 시군구=="충주시")
cheongju <- filter(data, 시군구=="청주시")
jecheon <- filter(data, 시군구=="제천시")

chungju
```

```r
p <- ggplot()
p
```

```r
p <- ggplot()
p <- p + geom_line(data= chungju, aes(발생월, 사고건수), color='red') + 
    geom_point(data=chungju, aes(발생월, 사고건수), color='red')
p
```

```r
p <- ggplot()

p <- p + geom_line(data = chungju, aes(x = 발생월, y = 사고건수), color = 'red') + 
         geom_point(data = chungju, aes(x = 발생월, y = 사고건수), color = 'red')
p <- p + geom_line(data = cheongju, aes(x = 발생월, y = 사고건수), color = 'blue') + 
         geom_point(data = cheongju, aes(x = 발생월, y = 사고건수), color = 'blue')
p <- p + geom_line(data = jecheon, aes(x = 발생월, y = 사고건수), color = 'green') + 
         geom_point(data = jecheon, aes(x = 발생월, y = 사고건수), color = 'green')
p
```

```r
library(dplyr)

cb <- data %>%
    filter(시도=='충북') %>%
    filter(grepl('시$', 시군구)) %>%
    select(발생월, 시군구, 사고건수)

head(cb)

table(cb$`시군구`)
```

```r
p <- ggplot(cb, aes(발생월, 사고건수,
    color=시군구, shape=시군구, linetype=시군구))
p + geom_point() + geom_line()
```

```r
gang <- data %>%
    filter(시도=='서울') %>%
    filter(grepl('^강', 시군구)) %>%
    select(발생월, 시군구, 사고건수)

head(cb)
```

```r
p <- ggplot(gang, aes(발생월, 사고건수,
    color=시군구, shape=시군구, linetype=시군구))
p + geom_point() + geom_line()
```

```r
p <- ggplot(gang, aes(발생월, 사고건수,
    color=시군구, shape=시군구, linetype=시군구))
p <- p + geom_point() + geom_line()
p <- p + facet_wrap(~시군구, nrow = 4) 
p
```
