```r
data <- read.csv("../mydata/도로교통공단_시군구별 월별 교통사고 통계_20231231.csv", fileEncoding = "CP949")
A <- subset(data, 시군구=="충주시")
B <- subset(data, 시군구=="청주시")
C <- subset(data, 시군구=="제천시")
```

```r
plot(A$`발생월`, A$`사고건수`, ylim=c(0,500), type='o', pch=20, col='red', 
      main="충북 3개 도시 월 교통사고 건수", xlab="월")
points(B$`발생월`, B$`사고건수`, ylim=c(0,500), type='o', pch=18, col='blue', 
      main="충북 3개 도시 월 교통사고 건수", xlab="월")      
points(C$`발생월`, C$`사고건수`, ylim=c(0,500), type='o', pch=17, col='green', 
      main="충북 3개 도시 월 교통사고 건수", xlab="월") 
```
