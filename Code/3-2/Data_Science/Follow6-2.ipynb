```r
library(dplyr)

data <- read.csv("../mydata/도로교통공단_시군구별 월별 교통사고 통계_20231231.csv", 
                 fileEncoding='CP949')
chungju <- filter(data, 시군구=="충주시")
cheongju <- filter(data, 시군구=="청주시")
jecheon <- filter(data, 시군구=="제천시")

chungju
```

```r
p <- ggplot()
p
```

```r
p <- ggplot()
p <- p + geom_line(data=chungju, aes(발생월, 사고건수), color='red') +
    geom_point(data=chungju, aes(발생월, 사고건수), color='red')
p
```

```r
p <- ggplot()
p <- p + geom_line(data=chungju, aes(발생월, 사고건수), color='red') + 
    geom_point(data=chungju, aes(발생월, 사고건수), color='red')
p <- p + geom_line(data=cheongju, aes(발생월, 사고건수), color='blue') + 
    geom_point(data=cheongju, aes(발생월, 사고건수), color='blue')
p <- p + geom_line(data=jecheon, aes(발생월, 사고건수), color='green') + 
    geom_point(data=jecheon, aes(발생월, 사고건수), color='green')
p
```

```r
p <- ggplot()
p <- p + geom_line(data=chungju, aes(발생월, 사고건수), color='red') + 
    geom_point(data=chungju, aes(발생월, 사고건수), color='red')
p <- p + geom_line(data=cheongju, aes(발생월, 사고건수), color='blue') + 
    geom_point(data=cheongju, aes(발생월, 사고건수), color='blue')
p <- p + geom_line(data=jecheon, aes(발생월, 사고건수), color='green') + 
    geom_point(data=jecheon, aes(발생월, 사고건수), color='green')
p
```

```r
options(repr.plot.width=7)
p <- ggplot()
p <- p + geom_line(data=chungju, aes(발생월, 사고건수, color='A', linetype="A")) + 
    geom_point(data=chungju, aes(발생월, 사고건수, color='A'))
p <- p + geom_line(data=cheongju, aes(발생월, 사고건수, color='B', linetype="B")) + 
    geom_point(data=cheongju, aes(발생월, 사고건수, color='B'))
p <- p + geom_line(data=jecheon, aes(발생월, 사고건수, color='C', linetype='C')) + 
    geom_point(data=jecheon, aes(발생월, 사고건수, color='C'))
p + scale_color_manual(values=c(A="red", B="blue",C="darkgreen"),
                      labels=c(A="충주시", B="청주시", C="제천시")) +
    scale_linetype_manual(values=c(A='solid', B='dashed', C='dotted'))
```

```r
library(readxl)
data1 <- read_xlsx("../mydata/기온통계-2024-1월-10월.xlsx", sheet=1)
data2 <- read_xlsx("../mydata/기온통계-2024-1월-10월.xlsx", sheet=2)
data3 <- read_xlsx("../mydata/기온통계-2024-1월-10월.xlsx", sheet=3)

data <- rbind(data1, data2, data3)
head(data)
```

```r
# 데이터 확인
table(data$`지점명`)

# 결측치 확인
any(is.na(data))
```

```r
# 데이터 전처리 - 연, 월, 일 분리
# 변수 '일' 은 숫자로 유형 변경
# 8월 데이터만 추출
library(dplyr)

month8 <- data %>%
	mutate(month=substr(일시, 6,7), day=substr(일시, 9,10))  %>%
    mutate(day2=as.numeric(day)) %>%
    filter(month=="08")
head(month8)
```

```r
# 기본 시계열 그래프
options(repr.plot.width=8)
p <- ggplot(month8, aes(day2, 최고기온, color=지점명))
p <- p + geom_point() + geom_line()
p
```

```r
# 제목 붙이기

p <- ggplot(month8, aes(day2, 최고기온, color=지점명))
p <- p + geom_point() + geom_line()
p <- p + labs(title="서울, 충주, 세종시의 2023년 8월 최고기온",
              x="일", y="최고기온")
p
```

```r
p <- ggplot(month8, aes(day2, 최고기온, color=지점명))
p <- p + geom_point() + geom_line()
p <- p + labs(title="서울, 충주, 세종시의 2023년 8월 최고기온",
              x="일", y="최고기온")
p + theme(plot.title=element_text(color="tomato", size=25),
          axis.title.x = element_text(size=30),
          axis.title.y = element_text(size=20, color='blue'))
```

```r
p <- ggplot(month8, aes(day2, 최고기온, color=지점명))
p <- p + geom_point() + geom_line()
p <- p + labs(title="서울, 충주, 세종시의 2023년 8월 최고기온",
              x="일", y="최고기온")
p <- p + theme(plot.title=element_text(color="tomato", size=25),
          axis.title.x = element_text(size=30),
          axis.title.y = element_text(size=20, color='blue'))
p + ylim(0, 40)
```

```r
p <- ggplot(month8, aes(day2, 최고기온, color=지점명))
p <- p + geom_point() + geom_line()
p <- p + labs(title="서울, 충주, 세종시의 2023년 8월 최고기온",
              x="일", y="최고기온")
p <- p + theme(plot.title=element_text(color="tomato", size=25),
          axis.title.x = element_text(size=15),
          axis.title.y = element_text(size=15, color='blue'))
p <- p +
    scale_x_continuous(breaks=1:31, labels=1:31) +
    scale_y_continuous(limits=c(0,40), breaks=seq(0,40,5), labels=seq(0,40,5))  
p
```
