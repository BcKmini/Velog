## 옵션 & layout

```r
library(readxl)
data1 <- read_xlsx("../mydata/기온통계-2024-1월-10월.xlsx", sheet=1)
data2 <- read_xlsx("../mydata/기온통계-2024-1월-10월.xlsx", sheet=2)
data3 <- read_xlsx("../mydata/기온통계-2024-1월-10월.xlsx", sheet=3)

data <- rbind(data1,data2,data3)
head(data)
```

```r
table(data$`지점명`)
```

```r
any(is.na(data))
```

```r
library(dplyr)

month8 <- data %>%
    mutate(월 = substr(일시, 6, 7), 
           일 = substr(일시, 9, 10)) %>% 
    mutate(day2 = as.numeric(일))

head(month8)
```

```r
library(ggplot2)
```

```r
# 기본 시계열 그래프
p <- ggplot(month8, aes(day2, 최고기온, color=지점명))
p <- p + geom_point() + geom_line()
p
```

```r
p <- ggplot(month8, aes(day2, 최고기온, color=지점명))
p <- p + geom_point() + geom_line()
p <- p + labs(title="서울, 충주, 세종시의 2023년 8월 최고기온", x="일", y="최고기온")
p
```
